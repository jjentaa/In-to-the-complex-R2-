```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1D random walk with 1 walker

N = 100 #number of step

pos_arr = np.zeros(N+1)
pos = 0 #starting pos of walker

pos_arr[0]=pos

#Moving right or left with equal prob
for i in range(N):
    r = np.random.rand()
    if r < 0.5:
        pos += 1 #right
    else:
        pos -= 1 #leftb
    pos_arr[i+1] = pos

plt.plot(pos_arr)
```

```python
#1D random walk with m walker

N = 10000 #num of step
M = 50 #num of walker

pos_arrary = np.zeros((M,N+1)) #m row n column

for j in range(M):
    pos = 0 #starting pos of walker
    pos_arrary[j, 0] = pos

    #Moving right or left with equal prob
    for i in range(N):
        r = np.random.rand()
        if r < 0.5:
            pos += 1 #right
        else:
            pos -= 1 #leftb
        pos_arrary[j, i+1] = pos
    plt.plot(pos_arrary[j][::10], linewidth = '1')

plt.show()
```

```python
#Basic statistic for random walk
# standard deviation(SD) - วัดการกระจายตัวของข้อมูล
# find S.D. for each walking step

sd_arr = np.zeros(N+1)

for k in range(N+1):
    sd_arr[k] = np.std(pos_arrary[:, k])

plt.plot(sd_arr, '.')
```

```python
data = pos_arrary[:,100]
hist, bins = np.histogram(data, bins=20, normed=True) #bins มีขอบออกมาด้วย จะเอาขอบมาเฉลี่ย เพื่อลดจำนวน
bins_left = bins[:-1] #left bin boundaries all element expect the last
bins_right = bins[1:]
bins_middle = 0.5*(bins_left+bins_right)
plt.plot(bins_middle, hist, label="step 100")
plt.legend()

data = pos_arrary[:,200]
hist, bins = np.histogram(data, bins=20, normed=True) #bins มีขอบออกมาด้วย จะเอาขอบมาเฉลี่ย เพื่อลดจำนวน
bins_left = bins[:-1] #left bin boundaries all element expect the last
bins_right = bins[1:]
bins_middle = 0.5*(bins_left+bins_right)
plt.plot(bins_middle, hist, label="step 200")
plt.legend()

data = pos_arrary[:,500]
hist, bins = np.histogram(data, bins=20, normed=True) #bins มีขอบออกมาด้วย จะเอาขอบมาเฉลี่ย เพื่อลดจำนวน
bins_left = bins[:-1] #left bin boundaries all element expect the last
bins_right = bins[1:]
bins_middle = 0.5*(bins_left+bins_right)
plt.plot(bins_middle, hist, label="step 500")
plt.legend()
```

```python
import numpy as np
x_data = np.arange(1, len(sd_arr)+1)
coefs = np.polyfit(x_data, sd_arr, 2)
x_model = np.linspace(x_data[0], x_data[-1])
poly = np.poly1d(coefs)
plt.plot(x_data, sd_arr, c="red", label="data")
plt.plot(x_model, poly(x_model), label="fit")
plt.xlabel('number of step')
plt.ylabel('S.D.')
plt.legend()
plt.grid()
```
