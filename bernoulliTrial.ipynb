```python
import random
import matplotlib.pyplot as plt
```

```python
def roll_dice():
    a = random.randint(1, 6)
    b = random.randint(1, 6)
    c = random.randint(1, 6)
    d = random.randint(1, 6)
    if(6 in [a, b, c, d]):
        return True
    else:
        return False
```

```python
pocket = 0
pockls = []
for i in range(10000):
    func = roll_dice()
    if(func):
        pocket -= 10
    else:
        pocket += 10

    pockls.append(pocket)
```

```python
plt.plot(pockls)
```

```python
expVal = pocket/10000
expVal
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bus = []
que = []

b=0
q=0
while q<10000:
    qq = random.uniform(0, 10)
    q += qq
    que.append(q)
while b<10001:
    bb = np.random.exponential(20)
    b += bb
    bus.append(b)
```

```python
waitls = []
index = 0
for i in que:
    while(i>bus[index]):
        index += 1
    wait = bus[index]-i
    waitls.append(wait)
```

```python
avg = sum(waitls)/len(waitls)
avg
```
